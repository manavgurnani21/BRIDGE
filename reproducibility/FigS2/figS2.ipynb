# Performance comparison
Here we walk through the necessary steps to reproduce the main figures of the performance comparison of different RBP datasets across cell lines using BRIDGE.

## 1. Load BRIDGE model and compute metrics
Please download the dataset from `Figshare`, unzip it, and place it in the expected directory (`BRIDGE/dataset/`). You also need to download the BRIDGE model and place it in the expected directory (`BRIDGE/model/`). Then run the code below to compute the BRIDGE metrics.

````python
%%bash
set -euo pipefail

ROOT_DIR="/home/wangyubo/code/BRIDGE"
cd "${ROOT_DIR}"
DATA_DIR="${ROOT_DIR}/dataset"
LOG="results/logs/validate_all.log"
mkdir -p results/logs

: > "$LOG"

for posfile in "$DATA_DIR"/*_pos.fa; do
  data_file="$(basename "$posfile")"
  data_file="${data_file%_pos.fa}"

  echo "Running validate for: ${data_file}" | tee -a "$LOG"

  python main.py \
    --validate \
    --data_path "$DATA_DIR" \
    --data_file "$data_file" \
    --device_num 0 \
    --Transformer_path ./RBPformer \
    --model_save_path ./results/model \
    2>&1 | tee -a "$LOG"
done
````

````python
%%bash
set -euo pipefail

ROOT_DIR="/home/wangyubo/code/BRIDGE"
cd "${ROOT_DIR}"
LOG="results/logs/validate_all.log"
OUT="results/benchmarks/BRIDGE_auc.tsv"
mkdir -p results/benchmarks

echo -e "RBP_Cell\tBRIDGE" > "$OUT"

grep -E '^[A-Za-z0-9]+_[A-Za-z0-9]+ auc:' "$LOG" \
  | awk '{print $1 "\t" $3}' \
  >> "$OUT" || true

echo "Wrote: $OUT"
cat "$OUT"
````

#### Load precomputed results for BRIDGE and other methods
```{note}
Running BRIDGE on the full dataset can take a long time. For convenience, we also provide the precomputed results for BRIDGE and all baseline methods on Figshare as `AUC_baselines.metrics`, `AUC_baselines.metrics`, and `MCC_baselines.metrics`, which you can download directly.
```

````python
%%R
base_path <- dirname(dirname(getwd()))
data <- read.table(
  paste0(base_path, "/results/metrics/AUC_baselines.metrics"),
  header = TRUE,
  sep = "\t",
  stringsAsFactors = FALSE
)
head(data)
````

#### Convert data format

````python
%%R
# Set row names as the first column
rownames(data) <- data[, 1]
data <- data[, -1]
mat <- as.matrix(data) # data to matrix
dim(mat)

# assign rownames as split variable
split <- as.factor(rownames(data))
````

**Color mapping**

````python
%%R
# Define a color ramp function for visualizing a matrix 'mat'
col_fun = colorRamp2(c(min(mat),  0.75, 0.81, 0.86, max(mat)),
                    c("#17336e","#b7d8eb", "#ffffff", "#fdfde8", "#fd3021"))
````

#### The plotting function

````python
%%R
# init circos plot
circos.clear()
# Set parameters for the circos plot
circos.par(
  canvas.xlim = c(-1.2, 1.2), # set x-axis range
  canvas.ylim = c(-1.2, 1.2), # set y-axis range
  clock.wise = FALSE, # set clockwise direction
  start.degree = 90, # set start angle
  gap.after = c(rep(1, length(rownames(data))-1), 25) # set gap between tracks
)

# plot heatmap for each column
# Loop through each column of the matrix 'mat' to create individual heatmaps in a circos plot
for (i in 1:ncol(mat)) {
  # Create a heatmap for the current column of 'mat' using the circos.heatmap function
  circos.heatmap(
    mat[, i, drop = FALSE], 
    split = split, 
    col = col_fun, 
    cluster = FALSE, 
    track.height = 0.07
  )
}

# draw legend
lgd <- Legend(
  at = c(0, 0.5, 1),
  labels = c("Low", "Medium", "High"),
  col_fun = colorRamp2(
    c(0, 0.5, 1),
    c("#17336e", "#ffffff", "#fd3021")
  ),
  title = "AUC",
  direction = "vertical",
  title_gp = gpar(fontsize = 10),
  labels_gp = gpar(fontsize = 8),
  legend_height = unit(3, "cm")
)

# set legend position and draw
draw(
  lgd,
  x = unit(0.95, "npc"),
  y = unit(0.15, "npc"),
  just = c("right", "bottom")
)

circos.clear()
````

#### ACC circos heatmap
This block generates `ACC plot` using `results/metrics/ACC_baselines.metrics`.

````python
%%R
# ---- ACC ----
base_path <- dirname(dirname(getwd()))

data_acc <- read.table(
  paste0(base_path, "/results/metrics/ACC_baselines.metrics"),
  header = TRUE,
  sep = "\t",
  stringsAsFactors = FALSE
)

# Set row names as the first column
rownames(data_acc) <- data_acc[, 1]
data_acc <- data_acc[, -1]

# data matrix
mat_acc <- as.matrix(data_acc)

# split factor
split_acc <- as.factor(rownames(data_acc))

# color mapping (ACC)
col_fun_acc <- colorRamp2(
  c(min(mat_acc), 0.75, 0.81, 0.86, max(mat_acc)),
  c("#17336e", "#b7d8eb", "#ffffff", "#fdfde8", "#fd3021")
)

circos.clear()
circos.par(
  canvas.xlim = c(-1, 1),
  canvas.ylim = c(-1, 1),
  clock.wise = FALSE,
  start.degree = 90,
  gap.after = c(rep(1, length(rownames(data_acc)) - 1), 25)
)

for (i in 1:ncol(mat_acc)) {
  circos.heatmap(
    mat_acc[, i, drop = FALSE],
    split = split_acc,
    col = col_fun_acc,
    cluster = FALSE,
    track.height = 0.07
  )
}

lgd_acc <- Legend(
  at = c(0, 0.5, 1),
  col_fun = colorRamp2(c(0, 0.5, 1), c("#6a7daa", "#ffffff", "#eb5d43")),
  title = "ACC",
  direction = "vertical",
  title_gp = gpar(fontsize = 10),
  labels_gp = gpar(fontsize = 8),
  legend_height = unit(3, "cm")
)

draw(
  lgd_acc,
  x = unit(0.95, "npc"),
  y = unit(0.15, "npc"),
  just = c("right", "bottom")
)

circos.clear()
````

#### MCC circos heatmap
This block generates `MCC plot` using `results/metrics/MCC_baselines.metrics`.
